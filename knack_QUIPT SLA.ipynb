```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
import unittest, time, re
import getpass
import datetime
from datetime import datetime, timedelta
import sys
import os
import zipfile
import glob

# Function to find the most recent file in the download folder
def get_most_recent_file(folder_path, timeout=30):
    start_time = time.time()
    while time.time() - start_time < timeout:
        files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if not f.endswith(".crdownload")]
        if files:
            most_recent_file = max(files, key=os.path.getmtime)  # Get the latest modified file
            return most_recent_file
        time.sleep(1)  # Wait a second before re-checking
    raise Exception("No file was downloaded within the timeout period.")
    
# Set up download folder
download_path = r"C:\Users\Alodia Bendicio\Downloads"

# DELETE FILES IF EXISTING
search_terms = ["QUIPT SLA Report"]
def delete_excel_files(search_terms, download_path):
    try:
        if not os.path.exists(download_path):
            print(f"Error: Downloads folder not found at {download_path}")
            return
        for term in search_terms:
            # Use glob with a wildcard to find Excel files (both .xls and .xlsx)
            pattern = os.path.join(download_path, f"*{term}*.xls*")  # Matches .xls, .xlsx, etc.
            files_to_delete = glob.glob(pattern)
            if files_to_delete:
                print(f"Found files containing '{term}':")
                for file_path in files_to_delete:
                    print(f"- {file_path}")
                    try:
                        os.remove(file_path)
                        print(f"  Deleted: {file_path}")
                    except OSError as e:
                        print(f"  Error deleting {file_path}: {e}")
            else:
                print(f"No files found containing '{term}'.")
    except Exception as e:
        print(f"An error occurred: {e}")

delete_excel_files(search_terms, download_path)

#Example Usage in Jupyter Notebook:
#After running the code above, if there is a file named "Camps DocsProdReport.xlsx" or "Hrs Docs Report.xls" in your downloads folder, these files will be deleted.

driver_path= r"C:\Program Files\Google\chromedriver-win64\chromedriver.exe"
service = Service(driver_path)
driver = webdriver.Chrome(service=service)
driver.implicitly_wait(30)
driver.verificationErrors = []
driver.accept_next_alert = True
driver.get("https://portal.knackrcm.com/User/login")
driver.find_element("id", "username").send_keys("*insert username*")
driver.find_element("id", "password").send_keys("*insert password*")
time.sleep(0.5)
driver.find_element(By.XPATH, "//button[@type='submit']").click()
driver.find_element(By.XPATH, "//nav[@id='page-leftbar']//span[text()='Reports']").click()
time.sleep(0.5)
driver.get("https://portal.healthybos.com/Report/SLA")
# driver.find_element(By.LINK_TEXT, "SLA").click()
time.sleep(0.5)


# Date Today
today = (datetime.now()).strftime('%Y-%m-%d')

try:
    QUIPT_SLA = fr"https://portal.knackrcm.com/Report/SLA.xlsx?offset=0&limit=1000000&Client%5B%5D=QUIPT&SubClient%5B%5D=&Status=&SubStatus=&SearchDateType=1&DateRangeStart=2025-05-01&DateRangeEnd={today}"
    driver.get(QUIPT_SLA) 
    print("Successfully Downloaded QUIPT SLA Report")
    time.sleep(30) # Wait for download before renaming
    sla_file = get_most_recent_file(download_path)
    new_sla_name = os.path.join(download_path, f"QUIPT SLA Report.xlsx")
    os.rename(sla_file, new_sla_name)
    print(f"Renamed QUIPT_SLA to {new_sla_name}")

except Exception as e:
   print(f"An error occurred: {e}")
    
today = (datetime.now()).strftime('%Y-%m-%d')
print(f"DONE EXPORTING " + today)

# driver.quit()  
```
