```python
# CODE FOR TUESDAY - FRIDAY (Yesterday's Data i.e. -1)

import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
import unittest, time, re
import getpass
import datetime
from datetime import datetime, timedelta
import sys
import os
import shutil
import glob
import ipywidgets as widgets
from IPython.display import display, clear_output

# Download Folder Path
download_path = os.path.expanduser(r"C:\Users\Alodia Bendicio\Downloads") # Expands to the user's Downloads directory
quipt_prod_path = os.path.expanduser(r"C:\Users\Alodia Bendicio\Downloads\QUIPT PROD CAMP FILES")  # Expands to the user's "held folder" directory.  Adjust if your path is different.

# DELETE ALL FILES IN QUIPT PROD CAMP FILES folder
search_terms = ["quiptprod"]
def delete_excel_files(search_terms, quipt_prod_path):
    try:
        if not os.path.exists(quipt_prod_path):
            print(f"Error: Downloads folder not found at {quipt_prod_path}")
            return
        for term in search_terms:
            pattern = os.path.join(quipt_prod_path, f"*{term}*.csv*")
            files_to_delete = glob.glob(pattern)
            if files_to_delete:
                print(f"Found files containing '{term}':")
                for file_path in files_to_delete:
                    print(f"- {file_path}")
                    try:
                        os.remove(file_path)
                        print(f"  Deleted: {file_path}")
                    except OSError as e:
                        print(f"  Error deleting {file_path}: {e}")
            else:
                print(f"No files found containing '{term}'.")
    except Exception as e:
        print(f"An error occurred: {e}")

delete_excel_files(search_terms, quipt_prod_path)

# Get Recent File to Rename
def get_most_recent_file(folder_path, timeout=30):
    start_time = time.time()
    while time.time() - start_time < timeout:
        files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if not f.endswith(".crdownload")]
        if files:
            most_recent_file = max(files, key=os.path.getmtime)  # Get the latest modified file
            return most_recent_file
        time.sleep(1)  # Wait a second before re-checking
    raise Exception("No file was downloaded within the timeout period.")
    
# XPATH
xpath = "//*[@id='DataTables_Table_0']/tbody/tr[1]/td[4]/a"

# CAMPAIGN Links
k = "https://manage.callshaper.com/CampaignExport/history/5742/8555"
l = "https://manage.callshaper.com/CampaignExport/history/5740/8557"
m = "https://manage.callshaper.com/CampaignExport/history/5743/8556"
n = "https://manage.callshaper.com/CampaignExport/history/5763/8743"


# CAMPAIGN NAMES
camp11 = "quiptprod Quipt All Calls"
camp12 = "quiptprod Quipt Edocs"
camp13 = "quiptprod Quipt PT Calls"
camp14 = "quiptprod Quipt Manual Calls"


try:
    driver_path= r"C:\Program Files\Google\chromedriver-win64\chromedriver.exe"
    service = Service(driver_path)
    driver = webdriver.Chrome(service=service)
    driver.implicitly_wait(30)
    driver.verificationErrors = []
    driver.accept_next_alert = True
# Enter the link to access and credentials
    driver.get("https://manage.callshaper.com/Report/Team")
    driver.find_element("id", "email").send_keys("*insert email")
    driver.find_element("id", "password").send_keys("*insert password*")
    time.sleep(0.5)
    driver.find_element(By.XPATH, "//button[@class='btn btn-inverse']").click()  # use XPATH -> add "//" -> change to single quotes "''" inside bracket
    time.sleep(0.5)
    print("Successful CDP Log In")
    
# CAMPAIGNS

# 11. Quipt All Calls
    driver.get(k)
    driver.find_element(By.XPATH, xpath).click()
    time.sleep(5) # Wait for download before renaming
    file_name = get_most_recent_file(download_path)
    new_file_name = os.path.join(download_path, camp11 + f".csv")
    os.rename(file_name, new_file_name)
    time.sleep(0.5)
    print(f"Succesfully downloaded '{camp11}'")

# Quipt Edocs
    driver.execute_script("window.open('');")
    driver.switch_to.window(driver.window_handles[1])
    print("Switch to the new tab")
    driver.get(l)
    driver.find_element(By.XPATH, xpath).click()
    time.sleep(5) # Wait for download before renaming
    file_name = get_most_recent_file(download_path)
    new_file_name = os.path.join(download_path, camp12 + f".csv")
    os.rename(file_name, new_file_name)
    time.sleep(0.5)
    print(f"Succesfully downloaded '{camp12}'")

# 13. Quipt PT Calls
    driver.execute_script("window.open('');")
    driver.switch_to.window(driver.window_handles[2])
    print("Switch to the new tab")
    driver.get(m)
    driver.find_element(By.XPATH, xpath).click()
    time.sleep(5) # Wait for download before renaming
    file_name = get_most_recent_file(download_path)
    new_file_name = os.path.join(download_path, camp13 + f".csv")
    os.rename(file_name, new_file_name)
    time.sleep(0.5)
    print(f"Succesfully downloaded '{camp13}'")
    
# 14. Quipt Manual Calls
    driver.execute_script("window.open('');")
    driver.switch_to.window(driver.window_handles[3])
    print("Switch to the new tab")
    driver.get(n)
    driver.find_element(By.XPATH, xpath).click()
    time.sleep(5) # Wait for download before renaming
    file_name = get_most_recent_file(download_path)
    new_file_name = os.path.join(download_path, camp14 + f".csv")
    os.rename(file_name, new_file_name)
    time.sleep(0.5)
    print(f"Succesfully downloaded '{camp14}'")


    print("Done Exporting 8 QUIPT Prod Files")


except Exception as e:
   print(f"An error occurred: {e}")
# finally:
#  driver.quit()

# MOVE FILES TO HELD PROD CAMP FILES FOLDER

# Use glob to find files containing "AH" (case-insensitive)
search_pattern = os.path.join(download_path, "*quiptprod*") # * is a wildcard
files_to_move = glob.glob(search_pattern)


if files_to_move:
    print(f"Found {len(files_to_move)} files to move:")
    for file_path in files_to_move:
        print(f" - {file_path}")

        try:
            # Extract just the filename for the destination path
            file_name = os.path.basename(file_path)
            destination_path = os.path.join(quipt_prod_path, file_name)

            # Move the file
            shutil.move(file_path, destination_path)
            print(f"Moved '{file_name}' to '{quipt_prod_path}'")

        except Exception as e:
            print(f"Error moving '{file_path}': {e}")

else:
    print("No files found matching the criteria.")

yesterday = (datetime.now() - timedelta(days=1)).strftime('%Y%m%d')
print(f"DONE " + yesterday)

```
