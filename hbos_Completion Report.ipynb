```python
# CODE FOR TUESDAY - FRIDAY (Yesterday's Data i.e. -1)

import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import unittest, time, re
import getpass
import datetime
from datetime import datetime, timedelta
import sys
import os
import zipfile
import glob

# Function to find the most recent file in the download folder
def get_most_recent_file(folder_path, timeout=30):
    start_time = time.time()
    while time.time() - start_time < timeout:
        files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if not f.endswith(".crdownload")]
        if files:
            most_recent_file = max(files, key=os.path.getmtime)  # Get the latest modified file
            return most_recent_file
        time.sleep(1)  # Wait a second before re-checking
    raise Exception("No file was downloaded within the timeout period.")
    
# Set up download folder
download_path = r"C:\Users\Alodia Bendicio\Downloads"

# DELETE FILES IF EXISTING
search_terms = ["DOCS Completes", "IV Completes"]
def delete_excel_files(search_terms, download_path):
    try:
        if not os.path.exists(download_path):
            print(f"Error: Downloads folder not found at {download_path}")
            return
        for term in search_terms:
            # Use glob with a wildcard to find Excel files (both .xls and .xlsx)
            pattern = os.path.join(download_path, f"*{term}*.xls*")  # Matches .xls, .xlsx, etc.
            files_to_delete = glob.glob(pattern)
            if files_to_delete:
                print(f"Found files containing '{term}':")
                for file_path in files_to_delete:
                    print(f"- {file_path}")
                    try:
                        os.remove(file_path)
                        print(f"  Deleted: {file_path}")
                    except OSError as e:
                        print(f"  Error deleting {file_path}: {e}")
            else:
                print(f"No files found containing '{term}'.")
    except Exception as e:
        print(f"An error occurred: {e}")

delete_excel_files(search_terms, download_path)
 
# Correct WebDriver Path
driver_path= r"C:\Program Files\Google\chromedriver-win64\chromedriver.exe"
# Setup ChromeDriver using Service (Selenium 4+)|
service = Service(driver_path)
driver = webdriver.Chrome(service=service)
# Define URLs and credentials
login_url = "https://manage.callshaper.com/Report/Team"
email = "insert email"  # Replace with your email
password = "insert password"  # Replace with your password

docs_completes = "https://manage.callshaper.com/CampaignExport/history/2089/7872"
iv_completes = "https://manage.callshaper.com/CampaignExport/history/2088/7873"
 
try:
# Step 1: Navigate to login page
    driver.get(login_url)
    print("Page Loaded for DOCS COMPLETES")
# Step 2: Wait for the login form to load
    WebDriverWait(driver, 10).until(
       EC.presence_of_element_located((By.ID, "email"))
    )
# Step 3: Enter credentials
    driver.find_element(By.ID, "email").send_keys(email)
    print("Entered email.")
    driver.find_element(By.ID, "password").send_keys(password)
    print("Entered password.")
# Step 4: Click the login button
    driver.find_element(By.XPATH, "//button[contains(@class, 'btn-inverse')]").click()
    print("Successfully Logged In")
# Step 5: For DOCS Completes
    driver.get(docs_completes)
    driver.find_element(By.XPATH, "//*[@id='DataTables_Table_0']/tbody/tr[1]/td[4]/a").click()  # use XPATH -> add "//" -> change to single quotes "''" inside bracket
    time.sleep(0.5)
    print("Succesfully downloaded DOCS Completes")
    # Wait for and rename DocsProd Report
    time.sleep(10) # Wait for download before renaming
    docs_file = get_most_recent_file(download_path)
    new_docs_name = os.path.join(download_path, f"DOCS Completes.xlsx")
    os.rename(docs_file, new_docs_name)
    print(f"Renamed DocsProd Report to {new_docs_name}")
# Step 5: For IV Completes
    driver.execute_script("window.open('');")
    driver.switch_to.window(driver.window_handles[1])
    print("Switch to the new tab")
    driver.get(iv_completes)
    print("Page Loaded for IV COMPLETES")
    driver.find_element(By.XPATH, "//*[@id='DataTables_Table_0']/tbody/tr[1]/td[4]/a").click()  # use XPATH -> add "//" -> change to single quotes "''" inside bracket
    time.sleep(0.5)
    print("Succesfully downloaded IV Completes")
    # Wait for and rename IVProd Report
    time.sleep(10) # Wait for download before renaming
    iv_file = get_most_recent_file(download_path)
    new_iv_name = os.path.join(download_path, f"IV Completes.xlsx")
    os.rename(iv_file, new_iv_name)
    print(f"Renamed IVProd Report to {new_iv_name}")
except Exception as e:
   print(f"An error occurred: {e}")

print("DONE")
```
