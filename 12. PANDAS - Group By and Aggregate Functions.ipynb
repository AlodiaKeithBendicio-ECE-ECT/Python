# GROUP BY and AGGREGATE FUNCTIONS

- group together the values in a column and display them all on the same row
- this allows to perform aggregate functions on those groupings

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\arfil\Downloads\DATA ANALYST FOLDER\JUPYTER NOTEBOOK\Flavors.csv")

df
```

#### Group By - Mean

```python
base_flavor_mean = df.groupby('Base Flavor').mean(numeric_only = True)

base_flavor_mean
```

#### Group By - Count

```python
base_flavor_count = df.groupby('Base Flavor').count()

base_flavor_count
```

#### Group By - Min

```python
base_flavor_min = df.groupby('Base Flavor').min()

base_flavor_min
```

#### Group By - Max

```python
base_flavor_max = df.groupby('Base Flavor').max()

base_flavor_max
```

#### Group By - Sum

```python
base_flavor_sum = df.groupby('Base Flavor').sum(numeric_only = True)

base_flavor_sum
```

#### Aggregation Function

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean', 'count', 'min', 'max', 'sum'], 
                              'Texture Rating':['mean', 'count', 'min', 'max', 'sum']})
```

#### Grouping on Multiple Columns

```python
df
```

```python
df.groupby(['Base Flavor', 'Liked']).mean(numeric_only = True)
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating':['mean', 'count', 'min', 'max', 'sum']})
```

#### Describe Function

```python
df.groupby('Base Flavor').describe()
```
