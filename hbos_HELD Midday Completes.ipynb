```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
import unittest, time, re
import getpass
import datetime
from datetime import datetime, timedelta
import sys
import os
import glob

# DELETE Files if Existing
download_path = r"C:\Users\Alodia Bendicio\Downloads"
search_terms = ["Team Report", "Campaign Report"]
def delete_excel_files(search_terms, download_path):
    try:
        if not os.path.exists(download_path):
            print(f"Error: Downloads folder not found at {download_path}")
            return
        for term in search_terms:
            # Use glob with a wildcard to find Excel files (both .xls and .xlsx)
            pattern = os.path.join(download_path, f"*{term}*.xls*")  # Matches .xls, .xlsx, etc.
            files_to_delete = glob.glob(pattern)
            if files_to_delete:
                print(f"Found files containing '{term}':")
                for file_path in files_to_delete:
                    print(f"- {file_path}")
                    try:
                        os.remove(file_path)
                        print(f"  Deleted: {file_path}")
                    except OSError as e:
                        print(f"  Error deleting {file_path}: {e}")
            else:
                print(f"No files found containing '{term}'.")
    except Exception as e:
        print(f"An error occurred: {e}")

delete_excel_files(search_terms, download_path)

#Example Usage in Jupyter Notebook:
#After running the code above, if there is a file named "Camps DocsProdReport.xlsx" or "Hrs Docs Report.xls" in your downloads folder, these files will be deleted.
# Function to find the most recent file in the download folder
def get_most_recent_file(folder_path, timeout=30):
    start_time = time.time()
    while time.time() - start_time < timeout:
        files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if not f.endswith(".crdownload")]
        if files:
            most_recent_file = max(files, key=os.path.getmtime)  # Get the latest modified file
            return most_recent_file
        time.sleep(1)  # Wait a second before re-checking
    raise Exception("No file was downloaded within the timeout period.")
try:
    driver_path= r"C:\Program Files\Google\chromedriver-win64\chromedriver.exe"
    service = Service(driver_path)
    driver = webdriver.Chrome(service=service)
    driver.implicitly_wait(30)
    driver.verificationErrors = []
    driver.accept_next_alert = True
# Enter the link to access and credentials
    driver.get("https://manage.callshaper.com/Report/Team")
    driver.find_element("id", "email").send_keys("insert email")
    driver.find_element("id", "password").send_keys("insert password")
    time.sleep(0.5)
    driver.find_element(By.XPATH, "//button[@class='btn btn-inverse']").click()  # use XPATH -> add "//" -> change to single quotes "''" inside bracket
    time.sleep(0.5)
# Set up download folder
    download_path = r"C:\Users\Alodia Bendicio\Downloads"
# Download Team Report
    driver.find_element(By.ID, "teamgroupby").click()
    driver.find_element(By.XPATH, "//option[text()='Team']").click()
    driver.find_element(By.ID, "run").click()
    time.sleep(0.5)
    driver.find_element(By.ID, "export").click()
    print("Succesfully downloaded Team Report")
# Wait for and rename Team Report
    time.sleep(10) # Wait for download before renaming
    team_file = get_most_recent_file(download_path)
    new_team_name = os.path.join(download_path, f"Team Report.xlsx")
    os.rename(team_file, new_team_name)
    print(f"Renamed Team Report to {new_team_name}")
# Download Campaign Report
    driver.find_element(By.ID, "teamgroupby").click()
    driver.find_element(By.XPATH, "//option[text()='Campaign']").click()
    driver.find_element(By.ID, "run").click()
    time.sleep(0.5)
    driver.find_element(By.ID, "export").click()
    print("Succesfully downloaded Campaign Report")
# Wait for and rename Campaign Report
    time.sleep(20) # Wait for download before renaming
    campaign_file = get_most_recent_file(download_path)
    new_campaign_name = os.path.join(download_path, f"Campaign Report.xlsx")
    os.rename(campaign_file, new_campaign_name)
    print(f"Renamed Campaign Report to {new_campaign_name}")
# DONE
    today = datetime.now().strftime('%Y%m%d')
    print(f"DONE EXPORTING " + today)
except Exception as e:
    print(f"An error occurred: {e}")
# finally:
#     driver.quit()
```
